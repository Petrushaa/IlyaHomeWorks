```python
def generate_pascal_triangle(levels):
    triangle = [[1]]
    current_level = [1]
    
    for level_num in range(1, levels):
        padded_level = [0] + current_level + [0]
        next_level = []
        

        for j in range(len(padded_level) - 1):
            next_level.append(padded_level[j] + padded_level[j + 1])
        
        current_level = next_level
        triangle.append(current_level)
    
    return triangle

def print_triangle(triangle):
    total_levels = len(triangle)
    
    for i, level in enumerate(triangle):

        formatted_numbers = [str(num) for num in level]
        level_string = ' '.join(formatted_numbers)
        

        indent = ' ' * (total_levels - i - 1)
        print(indent + level_string)

depth = int(input("Введите количество уровней треугольника: "))
pascal_triangle = generate_pascal_triangle(depth)
print_triangle(pascal_triangle)
```
