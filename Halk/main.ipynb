```python
print("Ты хочешь снять в кино'?")
if input().lower() == "да":
    print("В какой роли ты хочешь сняться? Халк или Локи?")
    if input().lower() == "халк":
        print("Сколько у тебя бицепс в объёме?")
        if int(input()) < 60:
            print("Иди домой лалка")
        else:
            print("Ты принят!")
    elif input().lower() == "локи":
        print('Кого ты больше любишь маму или папу?')
        if input().lower() == "маму":
            print("Иди домой огузок")
        elif input().lower() == "папу":
            print("Иди к папе и спроси кого больше любит")
        else:
            print("Ты с дет дома?")
    else:
        print('Такого персонажа нет')
elif input().lower() == "нет":
    print("покеда")
else:
    print('Ошибка пиши грамотнее')
```
