```python
N = int(input("Введи N: "))
print('Ты ввел:', N)

MY_CASH: dict[int, int] = {0: 0, 1: 1, 2: 1}
def fibonachi_cached(N: int) -> int:
    if N in MY_CASH:
        return MY_CASH[N]
    MY_CASH[N] = fibonachi_cached(N-1) + fibonachi_cached(N-2)
    return MY_CASH[N]

print('N-ое число в последовательности', fibonachi_cached(N))
```
