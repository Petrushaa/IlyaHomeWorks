```python
from rich import print as rprint

def greet(value: str):
    rprint(f'[bold magenta]Hello, {value}!')

def change_name(value: str) -> str:
    chars = list(value)
    for i in range(len(chars)):
        if i % 2 == 0:
            chars[i] = chars[i].upper()
    value = ''.join(chars)
    return value

print(change_name('lalka'))
greet('guy')

```
