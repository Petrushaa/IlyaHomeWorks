```python
from openai import OpenAI
from lecture_llm_api.settings import OpenAISettings
import dotenv
import json
from rich.console import Console
from rich.panel import Panel
from rich.markdown import Markdown

console = Console()

def print_chat(messages):
    for msg in messages:
        role = msg["role"]
        content = msg["content"]
        if role == "system":
            style = "bold blue"
            title = "[system]"
        elif role == "user":
            style = "bold green"
            title = "[user]"
        else:
            style = "bold magenta"
            title = "[assistant]"

        console.print(
            Panel(
                Markdown(content),
                title=title,
                style=style,
                border_style=style
            )
        )


dotenv.load_dotenv(dotenv.find_dotenv())

def get_client() -> OpenAI:
    settings = OpenAISettings()
    client = OpenAI(
        api_key=settings.openai_api_key.get_secret_value(),
        base_url=str(settings.openai_base_url),
    )
    return client

def save_history(filename:str, messages):
    with open(filename, "w", encoding="utf-8") as f:
        json.dump(messages, f, ensure_ascii=False, indent=4)
        
def load_history(filename):
    try:
        with open(filename, "r", encoding="utf-8") as f:
            return json.load(f)
    except FileNotFoundError:
        return []

def completions_variant(messages):
    client = OpenAI()
    completion = client.chat.completions.create(
        model="Qwen/Qwen3-Next-80B-A3B-Instruct",
        messages=messages,
    )
    messages.append(
        {
            "role": "assistant",
            "content": completion.choices[0].message.content,
        }
    )


if __name__ == "__main__":
    messages = load_history("messages.json")

    while True:
        inputMessage = console.input("[bold cyan]Ğ’Ğ²ĞµĞ´Ğ¸Ñ‚Ğµ ÑĞ¾Ğ¾Ğ±Ñ‰ĞµĞ½Ğ¸Ğµ:[/bold cyan] ")
        if inputMessage == "/exit":
            save_history("messages.json", messages)
            break
        elif inputMessage == "/clear":
            messages = [messages[0]]
            console.print("[bold red]Ğ˜ÑÑ‚Ğ¾Ñ€Ğ¸Ñ ÑĞ¾Ğ¾Ğ±Ñ‰ĞµĞ½Ğ¸Ğ¹ Ğ¾Ñ‡Ğ¸Ñ‰ĞµĞ½Ğ°[/bold red]")

        elif inputMessage.startswith("/system"):
            messages[0]["content"] = inputMessage[8:]
            console.print("[bold red]Ğ¡Ğ¸ÑÑ‚ĞµĞ¼Ğ½Ğ¾Ğµ ÑĞ¾Ğ¾Ğ±Ñ‰ĞµĞ½Ğ¸Ğµ Ğ¸Ğ·Ğ¼ĞµĞ½ĞµĞ½Ğ¾ Ğ½Ğ° [/bold red]" + f"[bold white]{inputMessage[8:]}[/bold white]")

        else:
            messages.append({"role": "user", "content": inputMessage})
            completions_variant(messages)
        print_chat(messages[-2:])
```
