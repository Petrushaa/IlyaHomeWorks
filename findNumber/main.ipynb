```python
print("укажи начало диапазона")
start = int(input())
print(start)
print("укажи конец диапазона")
end = int(input())
print(end)
print("Загадай число в диапазоне от", start, "до", end)

while True:
    mid = (start + end)//2
    print("Твое число < > = ", mid, "напоминаю границы региона:", start,end)
    ans = input()
    if ans == "<":
        print("<")
        end = mid
    elif ans == ">":
        print(">")
        start = mid
    elif ans == "=":
        print("=")
        print("Твое число", mid)
        break
    else:
        print(ans)
        print('Пиши грамотнее')

```
